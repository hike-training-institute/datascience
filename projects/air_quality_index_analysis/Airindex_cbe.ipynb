 #Indexing the REMARKS & AQI and Compute count of group from each 'STATION' in different areas  Using Groupby

```python
df_DCO238 = df[df['STATION']  == 'DCO - 238'].groupby(["Remarks"])[['AQI']].count().sort_values('AQI',ascending=False).reset_index()
#pt_DCO238=df_DCO238.sort_values('AQI',ascending=True).plot('Remarks', 'AQI', kind='barh', figsize=(5,5))
df_DCO238 
```
